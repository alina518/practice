```python
def emoji_to_text(text):
    if not text or not text.strip():
        return "Please enter a valid emoji or name!"

    result = demoji.replace_with_desc(text, " ").strip()

    # If no emoji was replaced, result may remain same as input
    if result == text.strip():
        return "Not a valid Emoji!"

    return result
# mtext='ğŸ˜€'
# emoji_to_text(mtext)
```

```python
def text_to_emoji(text):
    if not text or not text.strip():
        return "Please enter a valid emoji or name!"

    words = text.split()
    output_words = [custom_emoji_map.get(word.lower(), word) for word in words]

    # Check if no replacements happened
    if output_words == words:
        return "No emoji found for the above text!"

    return " ".join(output_words)
# my_text='I am happy and I think python is cool'
# emoji_text=text_to_emoji(my_text,custom_emoji_map)
# print(emoji_text)
```
