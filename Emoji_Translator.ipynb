```python
def emoji_to_text(text):
  return demoji.replace_with_desc(text,' ')
# mtext='ğŸ˜€'
# emoji_to_text(mtext)
```

```python
def text_to_emoji(text):
  words=text.split()
  output_words=[custom_emoji_map.get(word.lower(),word) for word in words]
  return ' '.join(output_words)
# my_text='I am happy and I think python is cool'
# emoji_text=text_to_emoji(my_text,custom_emoji_map)
# print(emoji_text)
```

```python
import gradio as gr
with gr.Blocks(title='Emoji Translator') as demo:
  gr.Markdown("Emoji <-> Name Translator")
  with gr.Tab('Emoji -> Text'):
    emoji_input=gr.Textbox(label='Emoji')
    emoji_output=gr.Textbox(label='Description',interactive=False)
    gr.Button('Convert').click(emoji_to_text,inputs=emoji_input,outputs=emoji_output)
  with gr.Tab('Text -> Emoji'):
    text_input=gr.Textbox(label='Text')
    text_output=gr.Textbox(label='Emoji',interactive=False)
    gr.Button('Convert').click(text_to_emoji,inputs=text_input,outputs=text_output)
demo.launch()
```
